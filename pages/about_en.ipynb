# 啥是创新产品研究员？
创新产品研究员有点像算法研究员，程序开发工程师，产品经理“三合一”的角色。这个角色的意义是，解决技术和算法落地难的问题。

尽管现在有很多超棒的信息技术和算法存在，例如大数据，人工智能，区块链等。但是如何利用好这些技术，解决行业，尤其是商业公司面临的实际业务问题，创造商业价值，依然是一个很大的难题。

创新产品研究员同时具备对技术和算法的应用能力，也熟悉自己领域的业务问题，因此可以更有效的设计产品解决方案；同时，创新产品研究员也有足够的开发实现能力，通过快速开发POC级别系统，验证自己的解决方案设计并且进行迭代改进，从而敏捷，低成本的找到解决方（或者快速失败，避免公司再错误的方向上投入）
